```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
df_tesla=pdr.get_data_yahoo('TSLA')
df_tesla
```

```python
#SIMPLE MOVING AVERAGE
```

```python
df_tesla['Open'].plot(figsize= (15,8))
```

```python
df_tesla['Open: 10 days rolling'] = df_tesla['Open'].rolling(window=10,min_periods=1).mean()
```

```python
df_tesla['Open: 10 days rolling']
```

```python
df_tesla[['Open','Open: 10 days rolling']].plot(xlim=['2020-01-01','2021-01-01'],figsize=(15,6))
```

```python
df_tesla['Open: 30 days rolling'] = df_tesla['Open'].rolling(window=30,min_periods=1).mean()
```

```python
df_tesla['Open: 40 days rolling'] = df_tesla['Open'].rolling(window=40,min_periods=1).mean()
```

```python
df_tesla[['Open','Open: 10 days rolling','Open: 30 days rolling','Open: 40 days rolling']].plot(xlim=['2020-01-01','2021-01-01'],figsize=(15,15))
```

```python
##CUMULATIVE MOVING AVG
```

```python
df_tesla['Open'].expanding().mean().plot(figsize=(10,6))
```

```python
# EWMA : Exponential weighted moving Average -> focus on recent data /current moving data
```

```python
## EMA TESLA SHARES
# Let smoothing factor be = 0.1
df_tesla['EMA_0.1'] = df_tesla['Open'].ewm(alpha=0.1,adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1']].plot(xlim=['2020-01-01','2021-01-01'],figsize=(15,5))
```

```python
df_tesla['EMA 5 days'] = df_tesla['Open'].ewm(span = 5).mean()
```

```python
df_tesla[['Open','EMA 5 days']].plot(figsize=(15,8))
```
